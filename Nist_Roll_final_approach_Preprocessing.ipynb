```python
roll_images = os.listdir(roll_folder)
len(roll_images)
```

```python
#step 2 resize image to 512-512

SIZE=512
for roll_image in roll_images:
    roll_image_path="/content/drive/MyDrive/roll/"+roll_image
    img = cv2.imread(roll_image_path)
    img = cv2.resize(img, (SIZE, SIZE))
    roll_save_path = os.path.join("/content/drive/MyDrive/Roll_folder_512_final/", roll_image)
    cv2.imwrite(roll_save_path, img)
```
